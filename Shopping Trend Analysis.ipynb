```python
arizona_sales = df[df['Location'] == 'Arizona'] 
print(arizona_sales)
```

```python
plt.figure(figsize=(10, 6))  
plt.title('Mostly Prepared Size by Genders in Arizona')
sns.countplot(x='Gender', hue='Size', data=arizona_sales)
plt.xlabel('Gender')  
plt.ylabel('Count')   
plt.legend(title='Size')  
plt.show()
```

```python
Top_five_items_purchased_Arizona_before= arizona_sales.groupby('Item Purchased')['Previous Purchases'].sum()  
```

```python
Top_fiveitems_purchased_Arizona_After= arizona_sales.groupby('Item Purchased')['Purchase Amount (USD)'].sum() 
```

```python
Top_five_before=Top_five_items_purchased_Arizona_before.head(5) 
print(Top_five_before)
```

```python
Top_five_present=Top_fiveitems_purchased_Arizona_After.head(5) 
print(Top_five_present) 
```

```python
plt.figure(figsize=(5, 5))
plt.title('Sales in Arizona Before and After the Promotion')
y_positions = np.arange(len(Top_five_before.index))
bar_width = 0.4 
plt.barh(y_positions - bar_width / 2, Top_five_before.values, height=bar_width, label='Before', color='red')
plt.barh(y_positions + bar_width / 2, Top_five_present.values, height=bar_width, label='After', color='skyblue')
plt.yticks(y_positions, Top_five_before.index)
plt.xlabel('Sales')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6,8))
plt.title('Seasonal Sales after and before the Promotion')
plt.bar(arizona_sales['Season'],arizona_sales['Purchase Amount (USD)'],label='After',color='orange') 
plt.bar(arizona_sales['Season'],arizona_sales['Previous Purchases'],label='Before',color='skyblue') 
plt.legend() 
plt.show();
```
